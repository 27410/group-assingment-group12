# Vitamin K2 production by _Lactococcus lactis_

## 1. Introduction

### 1.1 Literature review of the compound (<500 words)
Vitamin K2 also named menaquinone (MK), is a vitamin of the fat-soluble vitamin K family. The chemical backbone consist of a 2-methyl-1,4-naphthoquinone double-ring structure with different lipophilic side chains. One of its isomers, trans-isomer, has biological activity and plays an important role in the coagulation e.g. blood clotting, being a cofactor for the γ-glutamyl carboxylase enzyme, and calcium metabolism (Bøe & Holo, 2020a). Vitamin K2 was considered to have a potential for the Reference Daily Intake establishment due to its positive effect on reducing the risk of occurrence of cardiovascular diseases, bone loss, and, potentially, other age-related diseases (Maresz, 2021). 
Vitamin K2 is produced in the human body through the large intestine’s bacteria, like *Escherichia coli*. Therefore, its production is commercialized with liquid- and solid-state fermentations, with the former being preferred. Factors that affect the production yield are the microbial strain, substrate and its pre-treatment, temperature, and time of fermentation. For example, *Lactococcus lactis (L. lactis)* contributes to a great presence of vitamin K2 in cheese. *Bacillus subtilis (B. subtilis)* was studied to improve the production of the compound through different substrate usage. *Escherichia coli (E. coli)* was metabolically engineered to obtain the molecule (Bøe & Holo, 2020b) (Jadhav et al., 2022). 

As a product of bacterial metabolism, vitamin K2, can only be obtained from the large intestine or fermented products. Even though, more research should be done on the amount of vitamin presence in food products, it could be of great significance to secure the sufficient vitamin consumption from the diet or with dietary supplements (Zhou et al., 2022). Currently, there are several MK-7 products available on the market (e.g., Solaray, Terranova, or Now Foods). Moreover, vitamin K2 market is expected to grow, mainly because of increasing health awareness, more cases of identified deficiencies ect., stated by several market research companies (Market Research Ideas, 2023; POLARIS Market Research, 2020). This highlights the importance of studying the opportunities to increase the yield of vitamin K2 production with the help of bacteria. 
In *L. lactis*, MK is synthesized by two different pathways. Firstly, they can be formed from Acetyl-CoA through mevalonate and, subsequent, polyprenyl pathways, that result in the prenyl-diphosphate chain. Secondly, from phosphoenolpyruvate and D-erythrose-4-phosphate. Shikimate pathway leads to the formation of the chorismate, which by undergoing menaquinone pathway, forms 1,4-dihydroxy-2-napthoate (DHNA). The final molecules of the described pathways undergo several reactions to form menaquinones. It was observed that prenyl-diphosphate chain formed from Acetyl-CoA, rather than 1,4-dihydroxy-2-napthoate from D-erythrose-4-phosphate limits total menaquinone synthesis. Enzymes important for the mevalonate pathway are thiL, hmcM, mvaA, mvk, pmk and mvaD. For the polyprenyl pathway, fni, ispA, and preA. Chorismate enters the menaquinone pathway with the conversion by menF. The production of menaquinone is facilitated by the menA enzyme. In a study performed by Bøe & Holo, combined overexpression of mvk, preA, and menA increased menaquinone levels to a higher level than obtained individually (Bøe & Holo, 2020a).  


### 1.2 Literature review of the cell factory (<500 words)
*L lactis* is a gram positive, facultative anaerobic bacteria, lactic acid bacteria (LAB). While originally from plants *L. lactis* has since been used in the dairy industry for fermentation of cheese, yoghurt and other dairy products ​(Song et al., 2017)​. 
As a cell factory, *L. lactis* is well known, with a lot of studies describing its genome, promoters and expression systems. For secretion systems, *L. lactis* is gram positive bacteria and thereby more likely to secrete products extracellularly which makes recovery of the products easier and cheaper. Additionally, there is only one extracellular housekeeping protease in *L. lactis* which will decrease likelihood of product degradation. Furthermore, *L. lactis* has a “generally recognized as safe” (GRAS) status for many applications, which makes it much easier to commercialize the products where it is included ​(Song et al., 2017)​. 

*L. lactis* is a small microorganism in terms of genome and has 65% biased for AT base pairs, due to its low GC content. This is important when working with genes encoding a protein of interest, then this needs to be optimized for the codon usage. There is, additionally, a low cloning efficiency, which might be mitigated by using different strategies ​(Frelet-Barrand, 2022)​. Alternative cell factories would be other microorganisms with a GRAS level status. These could include *E. coli*, *Saccharomyces cerevisiae (S. cerevisiae)* and *B. subtilis*. *B. subtilis* is highly studied and optimized for K2 vitamin production already, with the production of the soybased product Natto ​(Kang et al., 2022)​. Other cell factories that naturally produce K2 vitamin is *Flavobacterium meningosepticum (F. meningosepticum)*, which is a gram negative, aerobic bacterium. *F. meningosepticum* is not a microorganism with a GRAS status, which would make it more difficult to commercialize. There are also not as many cell engineering tools for *F. meningosepticum* as there are for the beforementioned cell factories ​(Kang et al., 2022)​.

## 3. Selection and assessment of existing GSM (<500 words)

The chosen genome scale metabolic model (GSMM) is the international prototype for studying lactic acid bacterial genetics (Linares et al., 2010). It is the parent strain *Lactococcus lactis subsp. cremoris* MG1363 with the GSMM denoted iNF517. The genome-scale, constraint-based model accounts for 518 genes, 754 reactions and 650 metabolites. The model was experimentally validated and focused on improving the interplay between carbon and nitrogen, responsible for flavor formation in dairy products and production of flavor compounds (Flahaut et al., 2013). <br>

There are several lactic acid bacterial GSMM out there for different industrial applications, including both *Lactobacillus plantarum*, *Lactobacillus casei*, and *Lactococcus lactis* (Cunha et al., 2021; Smid et al., 2005; Teusink & Smid, 2006). The dairy field is well represented with a variety of well-studied organisms, both having genome data, transcriptomics, proteomic and metabolomics (Teusink & Smid, 2006). Further in recent years the development and construction of GSMM for improvement and assisted optimization of the industrial processes have been applied (Cunha et al., 2021; Smid et al., 2005). However, engineering complex biosynthetic pathways, such as for the vitamin production, requires genome-wide metabolic models (Smid et al., 2005). Which is why the well-studied, fully sequenced, and relatively simple *L. lactis* MG1363 was chosen. Further it has effectively developed genetic engineering tools, including cloning and expression systems that offer customizable features for L. lactis MG1363 (Song et al., 2017). <br>

The latest experimental validation was performed in 2013 with the focus on providing valuable insights into the metabolic networks underlying flavor formation in the dairy industry, based on interaction with the amino acid metabolism (Flahaut et al., 2013). They performed chemostat fermentations at different growth rates to incorporate the shift from acetogenic to homolactic fermentation. Further they quantified the unknown energy parameters for growth and non-growth maintenance (physiology at steady-state). They were also able to include nitrogen metabolism catabolic reactions and formation of volatiles in the metabolic network. However, the model does not contain lactose metabolism and proteinase production capacity (Flahaut et al., 2013). With the increasing interesting in using *L. lactis* as cell factory it nevertheless still seems possible to make reliable perditions based both on the GSMM (Flahaut et al., 2013) and the available publications focused on *L. lactis* MG1363 (Oliveira et al., 2005; Song et al., 2017). <br>

When running the memote, the annotation of genes is not matched to any database. However, with the overall good consistency, annotation of metabolites, reactions and Systems Biology Ontology (SBO), the model ends with a score of 85%.  There are two biomass reactions, so when performing engineering, this is important to be aware of. We are expecting to facilitate reliable predictions, however we need to be aware of the missing gene annotations if implementation of omics data is needed.

### 4.2 Model Validation

The initial model of *L lactis* was validated and enhanced using the KEGG database. This validation aimed to ensure that the model encompassed all relevant reactions and pathways critical for the accurate simulation of cellular metabolism. "Ubiquinone and other terpenoid-quinone biosynthesis" and "Terpenoid backbone biosynthesis" pathways from KEGG in *L. lactis* were used as references.

In addition to that, BiGG database was used. The distinction between using the BiGG and KEGG databases lies in their roles during the model enhancement process. The BiGG database primarily facilitates the engineering aspect of the model by adding reactions that are known to exist but might be missing in the initial model. On the other hand, the KEGG database contributes to completing the model by providing missing reactions and ensuring that the model encompasses a comprehensive representation of biological pathways. The ways of reactions addition are stated for each of the reactions.

### 4.3 Theoretical Yield and Productivity

When calculating the theoretical yield and productivity, it is performed for the original model and the model with the updated pathways.

In the context of the production of MK-7, the model initially predicted a production rate of 0. However, after the completion and validation process using KEGG, the model's capability improved significantly, resulting in a production rate of 0.04 mmol/gDW/h and a yield of 0.024 mmol MK-7/mmol glucose.

### 4.4 Medium Optimization

When optimizing Mk-7 production in *L. lactis*, the addition of glutamate and a subsequent incorporation of sucrose resulted in substantial improvements. The introduction of sucrose proved to be the most effective among iterated carbon sources.

Glutamate, an amino acid, contributes to both carbon and nitrogen metabolism, potentially influencing key pathways. Sucrose, a disaccharide, emerged as a superior carbon source, likely providing a favorable balance of carbon and energy for MK-7 production. These findings underscore the importance of specific substrates, such as glutamate and sucrose, in modulating the metabolic landscape of *L. lactis* for enhanced MK-7 production. ### 4.5 Single Gene Knockout

The influence of single gene deletion on the production of MK-7 was investigated. For that, every single gene in the GSM was deleted in isolation and the model objective function optimized.
The base production for comparison was 0.14914.
The figure below shows the production rate for each of the gene deletions. Not a single deletion lead to an increase in MK-7 production. Nonetheless, a set of essential genes (infeasible solution or productivity < 10 % of base productivity), non-influentaal genes (no change in productivity) and "genes of interest" (lowering producitivity when deleted) could be gained. The latter was identified as a interesting tool for the flux analysis, since this set of genes seems to somehow have an influence on productivity without being an essential part of the pathways leading to the target molecule.
 
For gene knockout analysis, it was decided to use glucose as a main energy source rather than sucrose, for which the optimal upper boundary had to be defined. It reflects conditions where the model represents a biologically relevant scenario.

### 4.6 Flux Variability Analysis

Flux Variability Analysis (FVA) of MK-7 production in the *L. lactis* GSM model revealed valuable insights into the metabolic behavior under various conditions. Initially, the model prioritized MK-7 production over biomass synthesis, with flexibility in glucose utilization and operating under anaerobic conditions. The addition of glutamate and sucrose led to a substantial increase in MK-7 production (from 0.04 to 11.1 mmol/gDW/h), accompanied increased ATP maintenance flux and biomass synthesis (from 0.004 to 0.1 mmol/gDW/h). Firstly, glutamate addition slightly enhanced productivity (from 0.04 to 0.15 mmol/gDW/h) and a further addeding sucrose resulted in MK-7 production from 0.15 to 11.1 mmol/gDW/h.

It was decided to use the extreme sucrose conditions, as it could reveal insights into the robustness and flexibility of the metabolic network, and better show the difference between pre- and post-sucrose addition condition. 

Attempts to optimize MK-7 production by introducing ATP and transitioning to aerobic conditions did not yield the desired improvements. The failure to enhance production with added ATP suggests a complex interplay of metabolic pathways, and the model's persistence in anaerobic mode indicates a robust preference for this mode of operation.

Furthermore, the observed change in the negative boundary of G3PD1ir, associated with NADH consumption, suggests improved cell functioning with the introduction of additional molecules.

### 4.7 Phenotypic Phase Plane (Production envelope)

So far, the model is solely optimized to maximize the production and thereby the yield of MK-7. However, the productivity optimized with the GSM is based on a single cell level. The productivity and yield of a bioprocess depends on the whole cell population. It is therefore necessary to balance the production of MK-7 be the individual cells and the cell growth of the population. On way to do that is by creating a production envelope, which show the operational space between productivity of MK-7 and biomass production. <br>
Below it can be seen that the relationship between MK-7 production and biomass production is not strictly linear. When operating at the points of maximum biomass production of 0.103, the productivity of MK-7 is still at 0.0913, which is still 61.2 % of the maximum MK-7 production. This offers a wide operational space with later production.

The model validation underscored the importance of accurate and comprehensive model representation and validation for reliable predictions. Overall, the combined use of both BiGG and KEGG databases served as a powerful approach for engineering and completing metabolic models, ensuring their fidelity in capturing the complexity of cellular metabolism.
